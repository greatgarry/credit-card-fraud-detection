```python
#loading dataset in pd df
cc_data = pd.read_csv("creditcard.csv")
```

```python
cc_data.drop_duplicates(inplace=True)
```

```python
cc_data.shape
```

```python
#seperating the data for analysis
legit=cc_data[cc_data.Class==0]
fraud=cc_data[cc_data.Class==1]
```

Under-Sampling

Building a sample dataset containing similar distribution of legit and fraud

No. of fraud transaction-> 492

```python
legit_sample = legit.sample(n=492)
```

```python
#Concatenating two df
new_df=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

Splitting data into features(X) and targets(y)

```python
X=new_df.drop(columns='Class',axis=1)
Y=new_df['Class']
```

```python
X
```

```python
Y
```

Split data into training and testing data

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

```python
#Logistic Regression
model = LogisticRegression()
```

```python
#training the model
model.fit(X_train,Y_train)
```

Evaluation

```python
#Accuracy Score(training)
X_train_predict=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_predict, Y_train)
```

```python
print('Accuracy on training data', training_data_accuracy)
```

```python
#Accuracy Score(test)
X_test_predict=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_predict, Y_test)
```

```python
print('Accuracy on testing data', testing_data_accuracy)
```
